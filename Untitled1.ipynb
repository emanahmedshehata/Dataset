```python
import pandas as pd

# Specify the filename of the uploaded dataset (e.g., 'books.csv')
uploaded_file = 'books.csv'  # Update with the actual filename

# Load the dataset using the filename of the uploaded file
df = pd.read_csv(uploaded_file)

# Display the first few rows of the dataset to verify
df.head()
```

```python
harry_potter_books = df[df['title'].str.contains('Harry Potter', case=False, na=False)]

# Check the filtered DataFrame to ensure it contains Harry Potter books
print(harry_potter_books)
```

```python
print(df.columns)
```

```python
# Check for missing values
print(df.isnull().sum())

# Handle missing values, if any
# For example, if 'original_publication_year' has missing values, you can fill them with the median year
median_year = df['original_publication_year'].median()
df['original_publication_year'].fillna(median_year, inplace=True)
```

```python
# Filter the dataset to focus only on the Harry Potter series
harry_potter_df = df[df['title'].str.contains('Harry Potter', case=False)]
# Sort the Harry Potter DataFrame by 'ratings_count' to find the most selling books
most_selling_books = harry_potter_df.sort_values(by='ratings_count', ascending=False)
```

```python
print("Most Selling Books in the Harry Potter Series:")
print(most_selling_books[['title', 'ratings_count']])
```

```python
# Calculate the average rating of the Harry Potter books
average_rating = harry_potter_df['average_rating'].mean()
```

```python
# Print the average rating of the Harry Potter books
print("Average Rating of Harry Potter Books:", average_rating)
```

```python
import matplotlib.pyplot as plt

# Plotting a histogram of ratings
plt.hist(harry_potter_df['average_rating'], bins=10, color='skyblue')
plt.xlabel('Average Rating')
plt.ylabel('Frequency')
plt.title('Histogram of Average Ratings for Harry Potter Books')
plt.show()
```
